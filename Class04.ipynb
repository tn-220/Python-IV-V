**Zip**

The zip() returns a zip object where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.



```python
numbers_1 = [1, 2, 3, 4, 5, 22]
numbers_2 = [5, 6, 7, 8, 9]

result = zip(numbers_1, numbers_2)
```

```python
print(list(result))

print(dict(result))
```

```python
myList = [(1, 5), (3, 7), (6, 4), (7, 9)]
#The * operator can be used in conjunction with zip() to unzip the list.
print(list(zip(*myList)))
print(myList)
```

```python
students = ["Luis", "Pedro", "Sara"]
midterm = [78, 80, 94]
final = [93, 88, 85]


finalGrades = [t for t in zip(students,midterm, final)]
print(finalGrades)
```

```python
finalGrades = {t[0]: (t[1], t[2]) for t in zip(students, midterm, final)}
print(finalGrades)
```

```python
finalGrades = {t[0]: max(t[1], t[2]) for t in zip(students, midterm, final)}
print(finalGrades)
```

```python
finalGrades2 = zip(students, map(lambda t: max(t[0], t[1]), zip(midterm, final)))

print(dict(finalGrades2))
```

**Make a dictionary with each person and the related average for the person**

```python
allAve = zip(students, map(lambda t: (t[0] + t[1])/len(t), zip(midterm, final)))

print(dict(allAve))
```

##Inheritance:
Inheritance allows us to define a class that inherits all the methods and properties from another class.

**Parent class** is the class being inherited from, also called base class.

**Child class** is the class that inherits from another class, also called derived class.




###Create a Parent Class
Any class can be a parent class, so the syntax is the same as creating any other class:



```python
#Create a class named Person, with firstname and lastname properties, and a printname method:

class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

#Use the Person class to create an object, and then execute the printname method:

x = Person("John", "Doe")
x.printname()
```

###Create a Child Class
To create a class that inherits the functionality from another class, send the parent class as a parameter when creating the child class:



```python
#Create a class named Student, which will inherit the properties and methods from the Person class:

class Student(Person):
  pass
```

```python
#Use the Student class to create an object, and then execute the printname method:

x = Student("Mike", "Olsen")
x.printname()
```

```python
class Pet:			               # Declare Pet class.		 
  def __init__(self, name):
    self.name = name


class Cat(Pet):			           # Cat class inherited from Pet class.
  def purr(self):			         # A member method specific to the Cat class.
    print(self.name + ' is purring...')	

class Dog(Pet):			           # Dog class inherited from Pet class.
  def bark(self):			         # A member method specific to the Dog class.	
	  print(self.name + ' is barking...')
```

```python
cat1 = Cat('Kitty')
dog1 = Dog('Fido')

cat1.purr()

dog1.bark()

```

###Add the \_\_init__() Function
A child class inherits the properties and methods from its parent.

We can add the \_\_init__() function to the child class (instead of the pass keyword).

When you add the \_\_init__() function, the child class will no longer inherit the parent's \_\_init__() function.




```python
#Add the __init__() function to the Student class:

class Student(Person):
  def __init__(self, fname, lname):
    pass
    #add properties etc.
```

To keep the inheritance of the parent's \_\_init__() function, add a call to the parent's \_\_init__() function:



```python
class Student(Person):
  def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)
```

###super() Function
Python also has a super() function that will make the child class inherit all the methods and properties from its parent:



```python
class Student(Person):
  def __init__(self, fname, lname):
    super().__init__(fname, lname)
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def sum(self):
        return (self.a+self.b)
    
    def subtract(self):
        return (self.a-self.b)
    
    def product(self):
        return (self.a*self.b)
    
    def divide(self):
        return (self.a/self.b)


calc1 = Calculator(5,3)
calc2 = Calculator(10,2)


print(calc1.sum())
print(calc1.subtract())
print(calc1.product())
print(calc1.divide())


print(calc2.sum())
print(calc2.subtract())
print(calc2.product())
print(calc2.divide())
```

##Quiz 0104 (Solution)

```python
class Aquarium:
  
    def __init__(self):
        self.animals = []
        print("An Aquarium has opened!")
    
    def enters(self, x):
        if not x in self.animals:
            self.animals.append(x)
            print( x + " is a new member of our Aquarium!")
        else:
            print("We already have " + x +".")
            
    def leaves(self, x):
        if x in self.animals:
            i = self.animals.index(x)
            del self.animals[i]
        else:
            print("No such sea creature in the Aquarium!")
            
    def show_all(self):
        print(self.animals)  
```

```python
my_aquarium = Aquarium()
```

```python
my_aquarium.enters('Turtle')
my_aquarium.enters('Dolphin')
my_aquarium.enters('Sea Lion')
my_aquarium.enters('Shark')
```

```python
my_aquarium.show_all()
```

```python
my_aquarium.leaves("Sea Lion")
my_aquarium.leaves("Killer Whale")
my_aquarium.show_all()
```

```python
my_aquarium.leaves("Turtle")
my_aquarium.show_all()
```

##Representation Method \_\_repr__
\_\_repr__is a special method used to represent a class's objects as a string. \_\_repr__ is called by the repr() built-in function. You can define your own string representation of your class objects using the \_\_repr__ method. Special methods are a set of predefined methods used to enrich your classes.

```python
class User:
  def __init__(self, name, family, age):
    self.name = name
    self.family = family
    self.age = age
  
  def __repr__(self):
    reprClass = str(self.name) + " " + str(self.family) + " is " + str(self.age)
    #print("#1 : ", reprClass)
    return (reprClass)

me = User('Pedro', 'Silva', 24)
you = User('sara', 'Smit', '20')

print(me)
print(you)

me.name
me.family
me.age
print(type(me))
```
